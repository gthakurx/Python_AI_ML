**Numeric Types Fundamentals: Integers, Floats, and Complex Number**


Integers in Python represent whole numbers without any fractional components. They are one of the most fundamental numeric data types in programming.




1.   Integers can be positive, negative, or zero (examples: 1234, 0, -55)
2.   In Python, integers have unlimited precision and can be of any size
3.   It can grow as large as your available memory (RAM) allows.
4.   Python stores large integers in chunks of 30 bits (on 64-bit systems).











| Language   | Type      | Size     | Range                  |
| ---------- | --------- | -------- | ---------------------- |
| **C**      | int       | 4 bytes  | 2 to 21          |
| **C**      | long long | 8 bytes  | 2 to 21          |
| **Python** | int       | variable | limited only by memory |






```python
# @title
import sys
sys.getsizeof(1)
```

```python
# @title
sys.getsizeof(-1) # Sign bit + magnitude chunks
```

```python
sys.getsizeof(10)
```

```python
sys.getsizeof(1000)
```

```python
sys.getsizeof(10**10)


# Memory layout of a Python integer (64-bit system)
# +---------------------+-------------+
# | Field               | Size (bytes)|
# +---------------------+-------------+
# | Ref count           | 8 bytes     |  # No.of variables reference this object
# | Type pointer -> int | 8 bytes     |  # Points to <class 'int'> type object
# | Object value header | 8 bytes     |  # Stores metadata for arb. integers
# | Actual value (5)    | 8 bytes     |  # Stores the integer value (small ints)
# +-----------------------+-----------+
# Total  32 bytes
```

**Floating-Point Numbers (float) in Python**


Floats in Python represent real numbers with decimal components.

1.   Floats store decimal values (e.g., 3.14, -0.001, 2.0)
2.   Python floats use 64-bit double-precision representation
3.   They occupy 8 bytes of memory
4.   Maximum value is approximately 1.8  10^308



```python
pi = 3.14
sys.getsizeof(pi)


#     +-------------------+
#     | Ref count (8B)    |  Reference: how many variables point to this object
#     | Type ptr (8B)     |  object class ( type of data)  <class 'float'>
#     | Value = 3.14 (8B) |  Stored as a C double
#     +-------------------+
#      Total = 24 bytes


```

**Boolean Logic in Python: Understanding `True/False` Values and Comparisons**

1. In Python, Boolean values are represented by the keywords True and False (capitalized)
2. The Python data type for Boolean values is bool
3. When converted to integers, True equals 1 and False equals 0

```python
var = False
variable = True
print(type(var))
print(type(variable))
```

**Boolean Expressions**



1.   A Boolean expression is any expression that evaluates to either True or False.

2.   These expressions are essential for decision-making in programs.

3.   This is done by comparison operators which compare values and return  Boolean results.

    *   == (equal to)
    *   != (not equal to)
    *   >(greater than)
    *   < (less than)
    *   >= (greater than or equal to)
    *   <= (less than or equal to)









```python
# comparison operators
x = 10
y = 20

print(x > y)
print(x < y)
print(x == y)
print(x != y)
print(x >= 10)
print(y <= 15)
```

**Logical Operators**

*   **and** :  Returns True only if both expressions are True
*   **or**  :   Returns True if either expression is True
*   **not** :  Negates a Boolean value (returns the opposite)







```python
a = True
b = False

print(a and b)
print(a or b)
print(not a)
```

**Python Arithmetic Operator Precedence**

1. Parentheses ()  Anything inside () is evaluated first.

2. Exponentiation **  Right-to-left associativity.

3. Unary + and -  For positive or negative numbers.

4. Multiplication *, Division /, Floor Division //, Modulus %  Left-to-right.

5. Addition + and Subtraction -  Left-to-right.


```python
# Parentheses
result = (2 + 3) * 4
print(result)
```

```python

result = 2 + 3 * 4
print(result)
```

```python
# Exponentiation
result = 2 ** 3 ** 2 # works from right to left

# first it will calculate 3**2 = 9
# second it will calculate 2**9 = 512
print(result)
```

```python
# unary operators

print(-(3 **2))  # exponentiation first and then unary
```

```python
-3 **(2 *2)
```

```python
-3**2*2-4
```

```python
print(10 * 2 // 4 % 3)

# left-to-right
# 10 * 2 = 20
# 20 // 4 = 5
# 5 % 3 = 2
```

```python
print(10 + 5 - 3)  # 12

# Left-to-right:
# 10 + 5 = 15  15 - 3 = 12
```
